```python
#FILE = "booking/json_data_raw.json"
#FILE = 'json_data_test.json'
#FILE= 'booking/json_data_1000_250.json'
FILE = 'json_data_test.json'

def handle_file(file_name="booking/listing-data_100_100.json"):
    data = open(file_name).read()
    data = data.replace("}{", "},{")
    
    data_arr = re.split("},{", data)
    data_arr[0] += "}"
    data_arr[0] = json.loads(data_arr[0])
    
    for i in range(len(data_arr)-2):
        data_arr[i+1] = "{" + data_arr[i+1] + "}"
        data_arr[i+1] = json.loads(data_arr[i+1])
        
    data_arr[len(data_arr)-1] = "{" + data_arr[len(data_arr)-1]
    data_arr[len(data_arr)-1] = json.loads(data_arr[len(data_arr)-1])

    return data_arr
```
